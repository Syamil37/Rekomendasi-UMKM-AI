```python
import pandas as pd


df_base = pd.read_csv('data_versi_6.csv')
df_base['kelurahan_upper'] = df_base['kelurahan'].str.upper().str.strip()

df_mini = pd.read_html('Data_Mini Market dan Pasar Swalayan - tabel.xls')[0]
df_apotek = pd.read_html('Daftar_Apotek di DKI Jakarta Tahun 2019 - tabel.xls')[0]

df_mini['kelurahan'] = df_mini['kelurahan'].str.upper().str.strip()
count_mini = df_mini.groupby('kelurahan').size().reset_index(name='Minimarket_Asli')

daftar_kel_valid = df_base['kelurahan_upper'].tolist()

def cari_kelurahan(alamat):
    alamat = str(alamat).upper()
    for kel in daftar_kel_valid:
        if kel in alamat:
            return kel
    return "TIDAK_KETEMU"

df_apotek['kelurahan_terdeteksi'] = df_apotek['alamat_apotek'].apply(cari_kelurahan)
count_apotek = df_apotek.groupby('kelurahan_terdeteksi').size().reset_index(name='Apotek_Asli')

df_final = pd.merge(df_base, count_mini, left_on='kelurahan_upper', right_on='kelurahan', how='left')
df_final = pd.merge(df_final, count_apotek, left_on='kelurahan_upper', right_on='kelurahan_terdeteksi', how='left')

df_final['Minimarket_Asli'] = df_final['Minimarket_Asli'].fillna(0)
df_final['Apotek_Asli'] = df_final['Apotek_Asli'].fillna(0)
df_final['Minimarket'] = df_final['Minimarket_Asli']
df_final['Apotek'] = df_final['Apotek_Asli']

kolom_sampah = ['kelurahan_upper', 'kelurahan_y', 'kelurahan_terdeteksi', 'Minimarket_Asli', 'Apotek_Asli']
df_final = df_final.drop(columns=[k for k in kolom_sampah if k in df_final.columns])

df_final.to_csv('data_versi_7_REAL.csv', index=False)
```

```python
import pandas as pd


df_base = pd.read_csv('data_versi_6.csv')

kolom_buang = ['Cafe', 'Laundry', 'Minimarket', 'Apotek']
df_base = df_base.drop(columns=[k for k in kolom_buang if k in df_base.columns])

df_base['kelurahan_upper'] = df_base['kelurahan'].str.upper().str.strip()

df_mini = pd.read_html('Data_Mini Market dan Pasar Swalayan - tabel.xls')[0]
df_apotek = pd.read_html('Daftar_Apotek di DKI Jakarta Tahun 2019 - tabel.xls')[0]

df_mini['kelurahan'] = df_mini['kelurahan'].str.upper().str.strip()
count_mini = df_mini.groupby('kelurahan').size().reset_index(name='Minimarket')

daftar_kel_valid = df_base['kelurahan_upper'].tolist()
def cari_kelurahan(alamat):
    alamat = str(alamat).upper()
    for kel in daftar_kel_valid:
        if kel in alamat: return kel
    return "TIDAK_KETEMU"

df_apotek['kelurahan_terdeteksi'] = df_apotek['alamat_apotek'].apply(cari_kelurahan)
count_apotek = df_apotek.groupby('kelurahan_terdeteksi').size().reset_index(name='Apotek')

df_final = pd.merge(df_base, count_mini, left_on='kelurahan_upper', right_on='kelurahan', how='left')
df_final = pd.merge(df_final, count_apotek, left_on='kelurahan_upper', right_on='kelurahan_terdeteksi', how='left')

df_final['Minimarket'] = df_final['Minimarket'].fillna(0)
df_final['Apotek'] = df_final['Apotek'].fillna(0)

kolom_sampah = ['kelurahan_upper', 'kelurahan_y', 'kelurahan_terdeteksi']
df_final = df_final.drop(columns=[k for k in kolom_sampah if k in df_final.columns])
if 'kelurahan_x' in df_final.columns:
    df_final = df_final.rename(columns={'kelurahan_x': 'kelurahan'})

df_final.to_csv('data_versi_8_PURE_REAL.csv', index=False)
```
